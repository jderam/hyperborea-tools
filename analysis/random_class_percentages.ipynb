```python
import pandas as pd
import matplotlib.pyplot as plt
from hyperborea.player_character import PlayerCharacter
```

```python
def get_class_counts(method: int = 3, subclasses: bool = True):
    generated_classes = []
    for i in range(100000):
        pc = PlayerCharacter(method=method, subclasses=subclasses)
        c = {
            "class_id": pc.to_dict()["class_id"],
            "class_name": pc.to_dict()["class_name"],
        }
        generated_classes.append(c)
    class_df = pd.DataFrame(generated_classes)
    class_count = class_df.groupby(class_df.class_name).count()
    return class_count
```

```python
method = 1
subclasses = False
c_counts = get_class_counts(method=method, subclasses=subclasses)
c_counts
```
